Polynomial Regression

y = b0 + b1X1 + b2X1^2 + ... + bnX1^n

It is still called Linear, because, y is still
linear function of the coefficients

linear and non-linear refers to the coefficients

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing Dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Training Linear

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

Training Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
print(X_poly)
```

Visualize

```python
plt.scatter(X, y, edgecolors='red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X, y, edgecolors='red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'orange')
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Higher Resolution and smoother curve

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'orange')
plt.title("Truth or Bluff (polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Predict

```python
lin_reg.predict([[6.5]])
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
